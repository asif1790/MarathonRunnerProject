```python
# COMBINE ALL FILTERS
```

```python
df2 = df[(df["Event distance/length"].isin(['50km','50mi'])) & (df["Year of event"]== 2018) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)== 'USA')] # filtering with 50mi/km and year 2017 and USA to new data frame
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0) # changing the main dataframe2 by assigning it.
```

```python
# clean athlete age by using 2017 - athlete year of birth
```

```python
df2['athlete age'] = 2018 - df2['Athlete year of birth'] # getting athlete approx age using 2017 - year of birth. assigning it to a new column variable.
```

```python
# remove h from athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0) # just removing h from the athlete performance to get the hours they ran the marathon in.
```

```python
#rename columns

#Year of event	int64
#Event dates	object
#Event name	object
#Event distance/length	object
#Event number of finishers	int64
#Athlete performance	object
#Athlete gender	object
#Athlete average speed	object
#Athlete ID	float64
#athlete age	int64
```

```python
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates': 'race_day',
                            'Event name': 'race_name',
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'race_number_of_finishers',
                            'Athlete performance': 'athlete_performance',
                            'Athlete gender': 'athlete_gender',
                            'Athlete average speed': 'athlete_average_speed',
                            'Athlete ID': 'athlete_id',
                            'athlete age': 'athlete_age'
})
```

```python
#season for the data -> Is summer slower than winter

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2
#split between two decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter' )
```

```python
df3.head()
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
